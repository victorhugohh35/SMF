```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf, erfinv

# Definir el rango de x
x = np.linspace(-3, 3, 500)
# Calcular y para la función compuesta
y = erfinv(1 - 2 * erf(x / np.sqrt(2)))

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='erfinv(1 - 2*erf(x/sqrt(2)))', color='green')
plt.title('Plot of erfinv(1 - 2*erf(x/sqrt(2)))')
plt.xlabel('x')
plt.ylabel('erfinv(1 - 2*erf(x/sqrt(2)))')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf, erfinv
from scipy.misc import derivative

# Definir la función compuesta
def composed_func(x):
    return erfinv(1 - 2 * erf(x / np.sqrt(2)))

# Definir el rango de x
x = np.linspace(-3, 3, 500)
# Calcular la derivada
y_derivative = np.array([derivative(composed_func, xi, dx=1e-6) for xi in x])

# Crear la gráfica
plt.figure(figsize=(10, 6))
plt.plot(x, y_derivative, label="Derivative of erfinv(1 - 2*erf(x/sqrt(2)))", color='purple')
plt.title("Derivative of erfinv(1 - 2*erf(x/sqrt(2)))")
plt.xlabel('x')
plt.ylabel("Derivative")
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.special import erf, erfinv
from scipy.misc import derivative

# Definir las funciones
def g(x):
    return erf(x)

def f(x):
    return erfinv(x)

# Definir las derivadas numéricas
def g_prime(x):
    return derivative(g, x, dx=1e-6)

def f_prime(x):
    return derivative(f, x, dx=1e-6)

# Rango de x para erfinv, que debe estar en [-1, 1]
x = np.linspace(-0.99, 0.99, 500)

# Calcular las derivadas
f_prime_values = np.array([f_prime(xi) for xi in x])
g_prime_values = np.array([g_prime(erfinv(xi)) for xi in x])

# Verificar la igualdad f'(x) = 1 / g'(f(x))
reciprocal_g_prime_values = 1 / g_prime_values

# Crear la gráfica
plt.figure(figsize=(12, 8))
plt.plot(x, f_prime_values, label="f'(x) = (erfinv(x))'", color='green')
plt.plot(x, reciprocal_g_prime_values, label="1 / g'(f(x)) = 1 / erf'(erfinv(x))", color='red', linestyle='dashed')
plt.title("Verificación de la igualdad f'(x) = 1 / g'(x)")
plt.xlabel('x')
plt.ylabel("Derivadas")
plt.legend()
plt.grid(True)
plt.show()
```
